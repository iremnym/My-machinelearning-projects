```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df=pd.read_csv('/kaggle/input/plane-price-prediction/Plane Price.csv')
df.head(3)
```

```python
df.info()
```

```python
df=df[["Rcmnd cruise Knots","Stall Knots dirty","Fuel gal/lbs","Eng out rate of climb","Price"]]
df.head(3)
df=df.dropna()
```

```python
y=df['Price']
x=df.drop("Price",axis=1)
```

* Normalizasyon Y ye yapılmaz!!
* Normalizasyon farklı sutunlardaki x değerlerini yakına aralıklara getirerek outlier etkisini azaltıyor (aykırı değer etkisi).
* Model performansımız da artar bu sayede.

```python
ss=StandardScaler()
x2=ss.fit_transform(x)
```

```python
x2=pd.DataFrame(x2)
x2.head(3)
```

```python
x2[0].mean()
```

```python
mm=MinMaxScaler()
x3=mm.fit_transform(x)
x3=pd.DataFrame(x3)
x3.head(3)
```

```python
x3[0].max(0)
```

```python
x3[0].min(0)
```

Bu demek ki 0 ile 1 arasında değerlere küçültür atar, bu da değerleri yaklaştırır. Yani bu normalizasyon yapmaktır.

```python
mm2=MinMaxScaler(feature_range=(0,10))
x4=mm2.fit_transform(x)
x4=pd.DataFrame(x4)
x4.head(3)
```
