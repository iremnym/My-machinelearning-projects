```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
df=pd.read_csv("/kaggle/input/insurance/insurance.csv")
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True)
```

**drop_first=True kodu**
sex_female sütunu silindi. (Çünkü "male" 1 ise, otomatik olarak "female" 0 olur.)
smoker_no sütunu silindi. (Çünkü "smoker_yes" 1 ise, otomatik olarak "smoker_no" 0 olur.)
region_west sütunu silindi. (Çünkü diğer 3 bölgeden biri 1 değilse, otomatik olarak west'tir.)
**Bu sayede çoklu doğrusal bağımlılık sorunu çözülmüş olur ve gereksiz bir sütun eklenmez.**

**drop_first=false kodu**
female, male, no yes burada hepsi var. false yazmadan çalıştırırsanda bu sonucu verir.

```python
df.head(3)
```

```python
y=df[['charges']]
x=df.drop('charges',axis=1)
```

**drop() fonksiyonu hem satır hem sütun silebilir.**
Eğer bir sütunu silmek istiyorsan axis=1 kullanmalısın.
Eğer bir satırı silmek istiyorsan axis=0 kullanmalısın.

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

**LinearRegression()** bir nesne oluşturur.
Bu nesne, y=aX + b şeklindeki doğrusal ilişkiyi öğrenmek için kullanılır.
Henüz veri ile eğitilmemiştir, sadece model oluşturulmuştur.

**fit(x, y)**, modelin X (bağımsız değişkenler) ile Y (bağımlı değişken) arasındaki ilişkileri öğrenmesini sağlar.
Model eğitilir.

```python
model.score(x,y)
```

```python
yeni_veri = pd.DataFrame([[19,26,0,1,1,0,0,1]], columns=x.columns)

# Tahmin yap
tahmin = model.predict(yeni_veri)

print("Tahmin Sonucu:", tahmin)
```

```python
df_error=pd.DataFrame()
df_error['y']=y
df_error
```

```python
y_tahmin=model.predict(x)
```

```python
df_error['tahmin']=y_tahmin
```

```python
df_error
```

```python
df_error.head(10)
```

```python
df_error['squared_error']=df_error['hata']**2
```

**MSE:** min square error, hataların karelerinin ortalamasıdır. Bunu yapma sebebimiz negatif, pozitif olmasıdır. Sönümlemenin önüne geçmektir.

```python
df_error['abs_error']=np.abs(df_error['hata'])
```

**MAE:** abstrack error, mutlak değer demek.

```python
df_error['percent_error']=np.abs(y-y_tahmin/y)
```

**MAPE******

```python
df_error.mean()
```
