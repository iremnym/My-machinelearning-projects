```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge
```

```python
df= pd.read_csv("/kaggle/input/student-scores/student_scores.csv")
df.head(3)
```

```python
y=df['Scores'] #1 boyutlu, 1 köşeli parantez
x=df[['Hours']] #2 boyutlu, 2 köşeli parantez
```


**DataFrame:** Bir 2 boyutlu veri yapısıdır ve genellikle tablo gibi düşünülebilir. Satırlar ve sütunlar içerir. Her sütun bir Series olabilir. Yani, DataFrame birden fazla Series'ten oluşur.

```python
lr=LinearRegression()
model=lr.fit(x,y)
model.score(x,y)
```

```python
r=Ridge(alpha=1)
modelr=r.fit(x,y)
modelr.score(x,y)
```

**Ridge Regresyonu ve Alpha (α):**

**Ridge Regresyonu:**
L2 regularizasyonu kullanır, modelin karmaşıklığını sınırlar.
Alpha (α):

Regularizasyon katsayısıdır, modelin parametrelerine ceza ekler.

**Ceza (Regularizasyon):**
Modelin parametrelerinin büyüklüğüne ceza ekler. Amaç, modelin aşırı karmaşık hale gelmesini engellemektir.
Ridge regresyonunda, parametrelerin karelerinin toplamına ceza eklenir.

**Alpha'nın Etkisi:**
Alpha Yüksekse: Model daha basit olur, daha fazla ceza.
Alpha Düşükse: Model daha karmaşık olur, daha az ceza.
Alpha = 0: Normal doğrusal regresyon, regularizasyon yoktur.
