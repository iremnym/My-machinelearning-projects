```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/kaggle/input/insurance/insurance.csv')
df.head(3)
```

```python
print(df.columns)
```

```python
df=pd.get_dummies(df, columns=['sex','smoker','region'],drop_first=True)
```

```python
df.head(3)
```

```python
df = df.astype(int)
```

```python
df.head(3)
```

```python
y=df['charges']
x=df.drop(columns=["charges"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=22,train_size=0.7)
```

```python
lr=LinearRegression()
model=lr.fit(x_train,y_train)
model.score(x_test,y_test)
```

**Random forest regressor=** ne kadar, kaç tane sorusunun cevabı

**Random forest classifier=** hangisi sorusunun cevabı

```python
rf=RandomForestRegressor(n_estimators=200)
#estimators agaç sayısı
model=rf.fit(x_train,y_train)
model.score(x_test,y_test)
```
